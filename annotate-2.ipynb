```python
import os
```

```python
import imgaug.augmenters as iaa
import imageio
import numpy as np

# Load your image
image = imageio.imread("image\sample garbage space.jpg")

# Define the directory where the augmented images will be saved
output_dir = "image\output2"

# Define the augmentation sequence
seq = iaa.Sequential([
    iaa.Fliplr(0.5), # horizontal flips
    iaa.Flipud(0.5), # vertical flips
    iaa.Affine(rotate=(-25, 25)), # rotate by -25 to +25 degrees
    iaa.Multiply((0.7, 1.3)), # change brightness, does not affect RGB channels
    iaa.LinearContrast((0.7, 1.3)), # improve or worsen the contrast
    iaa.AddToHueAndSaturation((-20, 20)) # change hue and saturation
])

# # Apply the augmentations
# images_aug = [seq(image=image) for _ in range(40)]

# # Save the augmented images
# for i, img in enumerate(images_aug):
#     imageio.imwrite(f"augmented_image_{i}.jpg", img)

# Generate and save augmented images
num_augmented_images = 40
for i in range(num_augmented_images):
    # Augment the image
    augmented_image = seq(image=image)
    
    # Save the augmented image
    output_path = os.path.join(output_dir, f"augmented_image_{i+1}.jpg")
    imageio.imwrite(output_path, augmented_image)

print(f"{num_augmented_images} augmented images have been saved to {output_dir}")
```
